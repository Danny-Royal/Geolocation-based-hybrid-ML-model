```python
# prompt: select the features that correlate the most with class and show

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the updated DataFrame
try:
    df = pd.read_csv('creditcard_with_ip.csv')
except FileNotFoundError:
    print("Error: 'creditcard_with_ip.csv' not found. Please run the previous code block to generate it.")
    exit()

# Select numerical features and the target variable ('Class')
numerical_cols = df.select_dtypes(include=['number']).columns
target_variable = 'Class'

# Calculate the correlation between numerical features and the target variable
correlations = df[numerical_cols].corrwith(df[target_variable])

# Sort correlations by absolute value to find the most correlated features
most_correlated_features = correlations.abs().sort_values(ascending=False)

# Print the most correlated features
print("Most Correlated Features with Class:\n", most_correlated_features)


# Create a heatmap of the correlations between the most correlated features and the target variable
plt.figure(figsize=(10, 6))
sns.heatmap(pd.DataFrame(most_correlated_features).transpose(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation with Class')
plt.show()
```

```python
# prompt: oversample the fraudulent cases to 9 times its size, undersample the legitimate cases to one third it size throughout the dataset

from sklearn.utils import resample

# Load the updated DataFrame
try:
    df = pd.read_csv('creditcard_with_ip.csv')
except FileNotFoundError:
    print("Error: 'creditcard_with_ip.csv' not found. Please run the previous code block to generate it.")
    exit()

# Separate majority and minority classes
df_majority = df[df.Class == 0]
df_minority = df[df.Class == 1]

# Upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True,     # sample with replacement
                                 n_samples=9 * len(df_minority),    # to match majority class
                                 random_state=123) # reproducible results

# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                   replace=False,    # sample without replacement
                                   n_samples=len(df_majority) // 3, # to one-third of its size
                                   random_state=123) # reproducible results

# Combine minority class with downsampled majority class
df_upsampled_downsampled = pd.concat([df_minority_upsampled, df_majority_downsampled])

# Display new class counts
print(df_upsampled_downsampled.Class.value_counts())
```

```python
# prompt: covert IP address to numbers

import pandas as pd

# Load the DataFrame (assuming it's already created and saved as 'creditcard_with_ip.csv')
try:
    df = pd.read_csv('creditcard_with_ip.csv')
except FileNotFoundError:
    print("Error: 'creditcard_with_ip.csv' not found. Please run the previous code block to generate it.")
    exit()

def ip_to_numeric(ip_address):
    parts = ip_address.split('.')
    numeric_ip = int(parts[0]) * (256**3) + int(parts[1]) * (256**2) + int(parts[2]) * 256 + int(parts[3])
    return numeric_ip

# Apply the conversion function to the 'ip_address' column
df['numeric_ip'] = df['ip_address'].apply(ip_to_numeric)

# Print the updated DataFrame with the new 'numeric_ip' column
print(df.head())
```

```python
# prompt: generate a hybrid logistic Regression, SVM and ANN

# Assuming X_train, X_test, y_train, y_test are already defined from previous code blocks.

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report
import tensorflow as tf
from tensorflow import keras
import numpy as np

# 1. Logistic Regression
log_clf = LogisticRegression(solver="lbfgs", random_state=42)
log_clf.fit(X_train, y_train)

# 2. Support Vector Machine (SVM)
svm_clf = SVC(gamma="scale", probability=True, random_state=42)  # probability=True for soft voting
svm_clf.fit(X_train, y_train)

# 3. Artificial Neural Network (ANN)
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1, verbose=0) # Set verbose to 0 to suppress output


# Create a VotingClassifier
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('svc', svm_clf)],
    voting='soft'  # Use soft voting (predict_proba)
)

voting_clf.fit(X_train, y_train)


# Evaluate the hybrid model
for clf in (log_clf, svm_clf, voting_clf):
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))

y_pred_ann = (model.predict(X_test) > 0.5).astype("int32")
print("ANN", accuracy_score(y_test, y_pred_ann))

# Example: Get predictions from the hybrid model
y_pred_voting = voting_clf.predict(X_test)

# Evaluate the hybrid model
print("Hybrid Model (VotingClassifier) Accuracy:", accuracy_score(y_test, y_pred_voting))
print(classification_report(y_test, y_pred_voting))

```
